```python
import requests
```

```python
import re
```

```python
pharm = requests.get('https://pharmacy.buffalo.edu/faculty-staff.html').text     # get faculty page in text
```

```python
pharm = pharm.replace('>', '>\n')      # replaces to make clearer
```

```python
#print(pharm)
```

```python
pattern = r'\bclass="facultyindexname" href=".*\b'       # pattern to find faculty links
```

```python
finds = re.findall(pattern, pharm)                       # find faculty links
```

```python
facultyLinks = []                                        # instiate new list for faculty
```

```python
for f in finds:
    d = f.split('class="facultyindexname" href="')       # loop through all links and split them
    facultyLinks.append(d[1])                            # appends to faculty list
    #print(d[1])                            # prints all the links
```

```python
facultyLinks       #list of faculty links
```

```python
def findInformation():
    for link in facultyLinks:                 # loops through all the faculty links
        if link != ('https://pharmacy.buffalo.edu/content/pharmacy/faculty-staff/faculty-profile.html?ubit=rominace'):
                    # bugged link doesn't work, faculty has no profile
            
            req = requests.get(link).text                   
            req = req.replace('>', '>\n')

            startName = '''<span id="lblFacultyName">'''                         # where to split from and to for name
            endName = '''<span class="facultyprofiledegrees">
<span id="lblFacultySuffix">'''

            facultyName_directory = req.split(startName)[1].split(endName)[0]     # splits req on startName and endName
            name_pattern = r'^[^<]*'                                              # pattern to find name
            name = re.findall(name_pattern, facultyName_directory)                # finds name and stores in name

            newString_Name = ''                                                   # new String
            facultyName = newString_Name.join(name[0].split('\n'))                # makes name be itself by splitting and
                                                                                  # joining, stores in facultyName
                
                                                                                  # where to split form and to for other info
            start = '''<div class="facultyprofile-expandablecontent">
<span id="lblAcademicOfficeAddressMobile"'''
            end = '''<div class="facultyprofile-expandablecontent">
</div>'''


            faculty_directory = req.split(start)[1].split(end)[0]              # splits req for the required info and stores 

            newString = ''                                                     # empty String

            phone_pattern = r'\bPhone: (.*)'                                   # pattern for phone number
            phone = re.findall(phone_pattern, faculty_directory)               # finds phone numbers
            newPhone = newString.join(phone).split('<br/>')[0]                 # splits <br/> from phone and makes it organized
            
            office_pattern = r'\w.+<br/>'                                      # pattern for office
            d = re.findall(office_pattern, faculty_directory)                  # finds office and all other information            
            empString = ''                                                     # empty string used to join next line
            everythingList = empString.join(d).split('<br/>')                  # splits on <br/> and joins to remove others
            addressList = []                                                   # creates new list to store address
            
            for count in range(len(everythingList)):                           # loops from 0 to length of everythingList
                phoneNumber = re.findall(phone_pattern, everythingList[count]) # finds phone pattern
                if len(phoneNumber) > 0:                                       # if there is phone pattern
                    addressList = everythingList[:count]                        # adds everything but phone pattern to list
                elif len(phoneNumber) == 0 and everythingList[count] != '':       # if not, then appends address to everythingList
                    addressList.append(everythingList[count])

            phoneString = ''                                                # new strings for later
            pString0_2 = ''
            pString3_5 = ''
            pString6_9 = ''
            for x in newPhone:                                  # loops through phone number and makes it all into only digits
                if x.isdigit() == True:
                    phoneString += x                            # puts in phoneString

            if len(phoneString) > 9:                            # if it is a real phone number then puts it into format () -
                pString0_2 = phoneString[0] + phoneString[1] + phoneString[2]
                pString3_5 = phoneString[3] + phoneString[4] + phoneString[5]
                pString6_9 = phoneString[6] + phoneString[7] + phoneString[8] + phoneString[9]


            # pNumber is phone number formatted
            pNumber = '({fThree}) {sThree}-{tFour}'.format(fThree = pString0_2, sThree = pString3_5, tFour = pString6_9)

            email_pattern = r'mailto:(\w+@buffalo.edu)'                       # pattern for email
            email = re.findall(email_pattern, faculty_directory)              # email[0] is the email

            if len(facultyName)>0:                                          # prints faculty name
                print(facultyName)
                
            if len(addressList) != 0:                                       # prints addreses
                for y in addressList:
                    print(y)
            else:
                print('No Address')
            
            if len(pNumber)>10:                                             # prints phone number
                print(pNumber)
            else: 
                print('No Phone')
            
            if len(email[0])>0:                                             # prints email
                print(email[0])
            
            else: 
                print('No E-mail')
            print('-'*20)                                                   # prints seperators between each person
```

```python
findInformation()                       # calls function, prints out information for each faculty
```
